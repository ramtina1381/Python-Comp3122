## 9- Model building

```python
fit_col = ['OverallQual', 'GrLivArea', 'GarageArea', 'Age', 'KitchenQualNum']
```

### Why we chose these features?

Since these features can be considered as the most efficient in case of hous assessment

#### Define your X_train, y_train, and X_test

```python
X = df[fit_col]
y = df['LogPrice']
X_test = df_test[fit_col]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

X_train = X_train.replace('NA', np.nan)  # Ensure 'NA' is replaced with np.nan
X_train = X_train.apply(pd.to_numeric, errors='coerce')  # Convert columns to numeric
X_train = X_train.fillna(X_train.mean())

X_test = X_test.replace('NA', np.nan)  # Ensure 'NA' is replaced with np.nan
X_test = X_test.apply(pd.to_numeric, errors='coerce')  # Convert columns to numeric
X_test = X_test.fillna(X_test.mean())
```

#### Instantiate and train a Linear Regression model

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

#### Predict the Price for the __training__ data and use it to compute fit quality using the RMSE metric

```python
from sklearn.metrics import mean_squared_error
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_error
import numpy as np
y_pred_training = model.predict(X_train)

rmse = np.sqrt(mean_squared_error(y_train, y_pred_training))
print("RMSE: %f" % (rmse))
```

#### One element of X_test.GarageArea is 'NA', replace it with `0` and make sure the column is numeric

```python
# Replace 'NA' with 0 in the 'GarageArea' column of X_test
X_test['GarageArea'] = X_test['GarageArea'].replace('NA', 0)

# Ensure the 'GarageArea' column is numeric (convert it to float or int)
X_test['GarageArea'] = pd.to_numeric(X_test['GarageArea'], errors='coerce')

# Check the result
print(X_test['GarageArea'].head())
```

```python
X_test['GarageArea'] = X_test['GarageArea'].apply(lambda x: 0 if x == 'NA' else x)
```

#### Predict the target (LogPrice) for the test data

```python
y_pred_test = model.predict(X_test)
rmse2 = np.sqrt(mean_squared_error(y_test, y_pred_test))  
print(f'RMSE for test data: {rmse2}')
```
